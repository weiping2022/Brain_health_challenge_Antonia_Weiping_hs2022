```python
com_col = ['Phase', 'RID', 'VISCODE2','PTID']
target = 'DIAG_GROUP'
```

```python
df = pd.read_csv('../data/main_file_final.csv').drop(['NPIKSEV'],axis=1) 
df = df[df['DIAG_GROUP'].notna()].reset_index(drop=True)  # keep the rows where DXCHANGE is not nan
df
```


#### we want to make the model stays simply, so we will use three binary models to answer two questions:
##### Q1: will the diagnosis get worse or not?
##### Q2: will the diagnosis get better or not?

```python
col_lst = [ 'DIAG_GROUP','NPIK1', 'NPIK2', 'NPIK3', 'NPIK4', 'NPIK5', 'NPIK6', 'NPIK7', 'NPIK8',
       'NPIK9A', 'NPIK9B', 'NPIK9C', 'insomnia','OSA']
df1 = df[com_col + col_lst].set_index(com_col).dropna(how='any',axis=0).reset_index()
col_lst_2 = [ 'DIAG_GROUP','NPIK1', 'NPIK2', 'NPIK3', 'NPIK4', 'NPIK5', 'NPIK6', 'NPIK7', 'NPIK8',
       'insomnia','OSA']




# Q1: will it get worse or not?
# two groups: 
    # MCI-AD or (MCI-MCI, MCI-NL); 
    # (NL-MCI, NL-AD) or NL-NL
label_map = { 'MCI-MCI':'MCI-MCINL','MCI-NL':'MCI-MCINL','MCI-AD':'MCI-AD','NL-AD':'NL-MCIAD','NL-MCI':'NL-MCIAD','NL-NL':'NL-NL'}
df3 = df1.copy()
df3['DIAG_GROUP'] = df3['DIAG_GROUP'].map(label_map)

# Q2: will it get better or not?
# two groups: 
    # MCI-NL or (MCI-MCI, MCI-AD); 
label_map2 = { 'MCI-MCI':'MCI-MCIAD','MCI-NL':'MCI-NL','MCI-AD':'MCI-MCIAD'}
df4 = df1.copy()
df4['DIAG_GROUP'] = df4['DIAG_GROUP'].map(label_map2)

df1
```

##### here I will use three self-generated functions (stored in dashboard_two.py)
   - usampling_scale_data(): under sampling, NOT SPLIT, scale data use standardscaling, pca transform data (explain 90% of the data variance)
   - cv_models(df,drop_lst,target,k): will return 168 models with cross-validation using 
        - three different input data: raw data, scaled data, pca-transformed data
        - four classification algorithms: logistic regression, decision tree, random forest, MLP, with the corresponding hyperparameter tuning
   - feature_importance(): analyse and plot by boxplot the feature importance from 10-fold cross validation 

```python
drop_lst = ['Phase', 'RID', 'VISCODE2', 'PTID',target]
k = 10
```

### model 1. will a patient with NL get worse or not?

```python
cv_models(g31,drop_lst,target,k)
```

```python
# 85trees. average weighted f1-score of 10-cross validation:0.608
n=85
res = usampling_scale_data(g31,drop_lst,target) 
X = res[0]  # unscaled input
y = res[3]
clf = RandomForestClassifier(n_estimators =n, random_state = 5862)
title_label = '{}-fold crossvalidation random forest ({} trees)'.format(k,n)
feature_importance(X,y,clf,k,title_label)
plt.suptitle('Target: NL-MCIAD: 63, NL-NL: 63')
#plt.xlabel('Sleep attributes',fontsize = 18)
plt.ylabel('Importance',fontsize = 22)
plt.xticks(fontsize=22, rotation=30)
plt.yticks(fontsize=18, rotation=30)

# plot the diagnosis distribution 
df_new = X.copy()
df_new[target] = y
f,axes = plt.subplots(nrows = len(lst)//2+1,ncols=2,figsize=(12,16))
axes = axes.ravel()
for i in range(len(lst)):
    ax = sns.countplot(data=df_new, hue=lst[i],x= target,ax = axes[i])
    sns.move_legend(ax, "upper right")
    ax.set_xticklabels(['NL - MCI/AD','NL - NL'])
    ax.set(ylabel=None)
    #ax.legend([],[], frameon=False)
```

### select data for image segmentation
- the patients with MCI are especially interesting for us. Because their AD status could get better, get worse, and not change.
- In all three sleep-Diag_changes models, NPIK9C (caregiver distress) is the most important factor to predict diagnosis changes. 
- So here I will select some observations, and further analyse the corresponding MRI data.

```python
df_img = df1[df1['DIAG_GROUP'].isin(['MCI-AD','MCI-MCI','MCI-NL'])& df1['VISCODE2']!='bl'] 
df_img.groupby(['NPIK9C','DIAG_GROUP']).count() 
df_img.info()
```

```python
img_info = pd.read_csv("../data/ADNI1_Complete_1Yr_1.5T_1_20_2022.csv") # the list of images
img_info=img_info.rename(columns = {'Subject':'PTID'})
img_info
```

```python
rus = RandomUnderSampler(random_state=432)
X_undersampled, y_unsampled = rus.fit_resample(df_img.drop(['NPIK9C'],axis=1), df_img['NPIK9C'])
df_img = pd.concat([X_undersampled, y_unsampled],axis=1)
df_img = df_img[df_img['PTID'].isin(img_info['PTID'].unique())].reset_index(drop=True) # filter out the observations where MRI are not available 
df_img
```

```python
df_img.groupby(['NPIK9C','DIAG_GROUP']).count() 
```
